```python
import pandas as pd
import calendar
from datetime import date
csv_path = "NJ Transit + Amtrak.csv"
```

```python
data = pd.read_csv(csv_path)

print(data)
```

```python
def order_by_scheduled_time(data: pd.DataFrame) -> pd.DataFrame: 
    copied_df = data.copy()
    sorted_df = copied_df.sort_values('scheduled_time')
    sorted_df.to_csv('sorted_df.csv', index=False)
    return sorted_df

data2 = order_by_scheduled_time(data)
```

```python
def drop_columns_and_nan(data2: pd.DataFrame) -> pd.DataFrame: 
    filtered_df = data2.copy()
    filtered_df = filtered_df.drop(['from', 'to'], axis=1)
    filtered_df = filtered_df.dropna()
    #filtered_df.to_csv('filtered_df.csv', index=False)
    return filtered_df

data3 = drop_columns_and_nan(data2)
    
```

```python
def convert_date_to_day(data3: pd.DataFrame) -> pd.DataFrame: 
    withdays_df = data3.copy()
    withdays_df['date'] = pd.to_datetime(withdays_df['date'])
    withdays_df['day'] = withdays_df['date'].dt.day_name()
    withdays_df = withdays_df.drop('date', axis=1)
    #withdays_df.to_csv('withdays_df.csv', index=False)
    return withdays_df

data4 = convert_date_to_day(data3)
```

```python
def convert_scheduled_time_to_part_of_the_day(data4: pd.DataFrame) -> pd.DataFrame: 
    withpartofday_df = data4.copy()
    withpartofday_df["hour"] = pd.to_datetime(withpartofday_df["scheduled_time"]).dt.hour
    withpartofday_df["part_of_the_day"] = withpartofday_df["hour"].apply(lambda x: 'late_night' if x < 4 else 'early_morning' if x < 8 else 'morning' if x < 12 else 'afternoon' if x < 16 else 'evening' if x < 20 else 'night')
    withpartofday_df = withpartofday_df.drop('hour', axis=1)
    #withpartofday_df.to_csv('withpartofday_df.csv', index=False)
    return withpartofday_df

data5 = convert_scheduled_time_to_part_of_the_day(data4)
```

```python
def convert_delay(data5: pd.DataFrame) -> pd.DataFrame: 
    withdelay_df = data5.copy()
    withdelay_df["delay"] = withdelay_df["delay_minutes"].apply(lambda x: '0' if x >= 0 and x < 5 else '1' if x >= 5 else '-1')
    #withdelay_df.to_csv('withdelay_df.csv', index=False)
    return withdelay_df

data6 = convert_delay(data5)
```

```python
def drop_unnecessary_columns(data6: pd.DataFrame) -> pd.DataFrame: 
    clean_df = data6.copy()
    clean_df = clean_df.drop(['train_id', 'scheduled_time', 'actual_time', 'delay_minutes'], axis=1)
    #clean_df.to_csv('clean_df.csv', index=False)
    return clean_df

data7 = drop_unnecessary_columns(data6)
```

```python
def save_first_60k(data7: pd.DataFrame, to_csv_path: str) -> pd.DataFrame: 
    copy_df = data7.copy()
    first60k_df = copy_df.iloc[:60000].copy()
    first60k_df.to_csv(to_csv_path, index=False)
    return first60k_df

data8 = save_first_60k(data7, "first60k_df.csv")
```
